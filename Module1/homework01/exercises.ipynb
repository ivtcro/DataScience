## Numpy

Импортировать NumPy под именем np

```python
import numpy as np
print (np.__version__)
```

Создать вектор (одномерный массив) размера 10, заполненный нулями

```python
a = np.array([0]*10)
b = np.zeros(10,dtype=int)
print(a, b)
```

Создать вектор размера 10, заполненный единицами

```python
a = np.array([1]*10)
b = np.ones(10,dtype=int)
print(a, b)
```

Создать вектор размера 10, заполненный числом 2.5

```python
a = np.array([2.5]*10)
b = np.full(10,2.5)
print(a, b)
```

Создать вектор размера 10, заполненный нулями, но пятый элемент равен 1

```python
b = np.zeros(10,dtype=int)
b[4]=1
print(b)
```

Развернуть вектор (первый становится последним)

```python
a = np.arange(10)
print(np.flip(a))
print(a[::-1])
```

Создать матрицу (двумерный массив) 3x3 со значениями от 0 до 8

```python
a = np.fromfunction(lambda x,y:x*3+y, (3,3), dtype=int)
b = np.arange(9).reshape(3,3)

print(a)
print(b)
```

Дан массив (10,2) координат, найти расстояние от каждой точки до каждой

```python
from scipy.spatial import distance

a = np.random.randint(1,100,(10,2))
print (a)
b = distance.cdist(a,a)
print(b)
```

Случайно расположить p элементов в 2D массив

```python
p = 10
# 2D array size
q = (10,10)
# array with p elements
a = np.random.randint(1,100,p)
# 2d array
b = np.zeros(q)
indices_to_change=np.random.choice(range(q[0]*q[1]), p, False)
np.put(b, indices_to_change, a)
print (b)
```

Дан вектор [1, 2, 3, 4, 5], построить новый вектор с тремя нулями между каждым значением

```python
mfactor = 3
a = np.array([1, 2, 3, 4, 5])
b = np.zeros(len(a) * mfactor - (mfactor - 1), dtype=int)
print(a)
print(b)
b[0::mfactor]=a
print(b)
```

Найти наиболее частое значение в массиве

```python
a = np.random.randint(0,10,10)
print (a)
print (np.bincount(a).argmax())
```

Найти n наибольших значений в массиве

```python
n=5
a = np.random.randint(1,10000,1000)
print (a)
print(np.sort(a)[-5:])
```

## Pandas - фильтрация данных

Считать данные по  [адресу](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 
Присвоить переменной chipo

```python
import requests

url="https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url, sep='\t', converters={"item_price":lambda x: float(x.replace('$', ''))})

chipo.head(10)


```

Как много продуктов стоят более чем $10.00?

```python
chipo[chipo['item_price'] > 10].shape[0]
```

Какова цена каждого предмета? Напечатать столбцы item_name и item_price

```python
chipo[['item_name', 'item_price']]
```

Отсортировать по имени предмета

```python
chipo.sort_values(by=['item_name'])
```

Сколько было заказано самого дорогого предмета?

```python
chipo[chipo['item_price'] == chipo['item_price'].max()]['quantity'].sum()
```

Сколько раз был заказан предмет "Veggie Salad Bowl"?

```python
chipo[chipo['item_name'] == "Veggie Salad Bowl"]['quantity'].count()
```

Сколько раз люди заказали более чем 1 предмет "Canned Soda"?

```python
chipo[(chipo['quantity'] > 1) & (chipo['item_name'] == "Canned Soda")].shape[0]
```
